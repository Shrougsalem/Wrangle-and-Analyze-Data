# Analysis and Insight Report

In this project we gathered Real-world data from a variety of sources and in a variety of formats. The first file is the tweet archive of Twitter user @dog_rates, also known as WeRateDogs, an additional data from the Twitter API (JSON file) and finally the image predictions file.  The quality and tidiness issues are assessed, then every issue is cleaned using define, code, test process. This is called data wrangling. The wrangling process is documented. This report communicate findings using the analysis and visualizations that we did.
After exploring, analyzing the data we found that:
### The tweets amount over the time

After a small peak in the number of tweets in 2015, we notice that the number has drastically decreased in 2016. (over 300 tweets/month to reach 10 tweets/month). 
### The top twitter sources

98.05% of the tweets were from Twitter for iPhone. 1.43% of the tweets are from Twitter Web Client, also, 0.51% of tweets are from TweetDeck  the least source of tweets).
### Ratings

We saw the rating distribution. In over 400 tweets the rating was 12/10. 10 and 11 are close after. The average rating is 10.24. 
### Top dog names

Charlie is one of the most common dog name with 11 tweets. Followed by Cooper, Lucy and Oliver.
### Common dog stage

In 10.25% of the tweets the dogs were in the pupper stage. The least common stage is floofer.
### Common breed

Here in the top 10 breeds, the most common breed (tweeted about) is the golden retriever with over 130 tweets. Then we have Labrador retriever with over 90 tweets followed by the Pembroke with 90 tweets.
